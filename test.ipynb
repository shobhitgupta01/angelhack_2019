```python
from bs4 import BeautifulSoup

with open("./new_topics.html") as fp:
    soup = BeautifulSoup(fp,'html.parser')
```

```python
import re
```

```python
re.sub(r'[^a-zA-Z ]+', '', '200 happy t00 go 129.129//')
```

```python
topic_list=[]
for link in soup.find_all('a'):
    s = link.text
    s = re.sub(r'[^a-zA-Z ]+', '', s)
    if len(s)>2:
        topic_list.append(s.strip())
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(topic_list,columns=["topics"])
```
